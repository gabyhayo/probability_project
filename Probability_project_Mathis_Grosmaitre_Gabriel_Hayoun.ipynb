# Une construction alternative

On considère maintenant la construction suivante, définie pour tout $t \in [0,1]$:
$$\tilde{Z}^{(n)}_t = \frac{1}{\sqrt{n}}\sum_{i=1}^n X_i (1_{[0,U_i]}(t)-1_{]U_i,1]}(t))$$
où les $X_i$ sont définis tels que précédemment et les $U_i$ sont des variables aléatoires i.i.d. de loi uniforme sur $[0,1]$.
Cette construction permet de définir un algorithme alternatif de simulation d'une (approximation de) trajectoire brownienne, cette fois-ci non standard au sens où on n'a pas systématiquement $\tilde{Z}^{(n)}_0 = 0$.

1. Discrétiser l'intervalle $[0,1]$ en 1000 points de manière régulière et implémenter l'algorithme (commencer par n = 1). Représenter une trajectoire pour $n = 1000$.

```python
n = 1000
Z = []
Y = 0
for i in range(n+1) :
    X = np.random.choice([-1,1])
    U = np.random.rand()
    if i/n <= U :
        Y += X
    else :
        Y += -X
    Z.append(Y)
Z = Z / np.sqrt(n)
plt.plot(range(n+1), Z, color = "crimson")
plt.title("Simulation de Zt(n) alternatif aux points ti au fur et à mesure des itérations")
plt.xlabel("Itération")
plt.ylabel("Z")
plt.grid()
plt.show()
```

2. Calculer l'espérance et la variance de $\tilde{Z}^{(n)}_t$. En déduire que $\forall t \in [0,1],~ \tilde{Z}^{(n)}_t\xrightarrow[n \to \infty]{\mathcal{L}}\mathcal{N}(0,1)$. Le vérifier expérimentalement aux points $\{0,0.25,0.5,0.75,1\}$ sur la base d'un échantillon de 1000 réalisations de $\tilde{Z}^{(n)}$.

### Q2
Soit $t \in [0,1]$\
Soit $ n \in \mathbb{N}$\
Soit $ i \in [\![0,n]\!]$
$\newline$Notons $W_i^t = X_i (1_{[0,U_i]}(t)-1_{]U_i,1]}(t))$. \
Donc : 
$
\tilde{Z}_t^{(n)} = \frac{1}{\sqrt{n}}\sum_{i=1}^n W_i^t
$

On a que $W_i^t(\Omega) = \{-1, 1\}$ et que  $\mathbb{P}(W_i^t = 1) = \mathbb{P}(W_i^t = -1) = \dfrac{1}{2}$

Ainsi :
$$
\mathbb{E}(W_i^t) = 1\times\frac{1}{2} - 1\times\frac{1}{2}=0
$$
On a donc par linéarité de l'espérance :
$$
\mathbb{E}(\tilde{Z}_t^{(n)})=\mathbb{E}(\frac{1}{\sqrt{n}}\sum_{i=1}^n W_i^t)=\frac{1}{\sqrt{n}}\sum_{i=1}^n \mathbb{E}(W_i^t)=0
$$
i.e
$$
\boxed {\mathbb{E}(\tilde{Z}_t^{(n)})=0}
$$
et 
$$
\mathbb V(W_i^t) = 1^2 \times \dfrac{1}{2} + (-1)^2 \times \dfrac{1}{2} = 1
$$

Commes les $W_i^t$ sont indépendantes par indépendance des $X_i$ et des $U_i$,on a :
$$
\mathbb{V}(\tilde{Z}^{(n)}_t) = \mathbb{V}(\frac{1}{\sqrt{n}}\sum_{i=1}^n W_i^t) = \frac{1}{n}\sum_{i=1}^n \mathbb{V}(W_i^t) = 1
$$
Ainsi :
$$
\boxed {\mathbb{V}(\tilde{Z}^{(n)}_t)=1}
$$

```python
n = 4
liste_Z = []
taille_echantillons = 1000
for ech in range(taille_echantillons) :
    Z = 0
    for i in range(n+1) :
        X = np.random.choice([-1,1])
        U = np.random.rand()
        if i/n <= U :
            Z += X
        else :
            Z += -X
    Z = Z / np.sqrt(n)
    liste_Z.append(Z)
```

Posons $T_i^h = W_i^t W_i^{t+h} = X_i^2(1_{[0,U_i]}(t)-1_{]U_i,1]}(t))(1_{[0,U_i]}(t+h)-1_{]U_i,1]}(t+h))$

$$

$$
\mathbb{P}(T_i^h = -1)= 1-\mathbb{P}(T_i^h = 1)=h
$$
On en déduit alors que :
$$
\begin{align}
\mathbb{E}(T_i^h) &= 1 \times (1-h) + (-1)\times h \\

\end{align}
$$

$$
\begin{align}
\mathbb{E}((\tilde{Z}^{(n)}_t-\tilde{Z}^{(n)}_{t+h})^2) &= 1  -2\frac{1}{n}\sum_{i=1}^n \mathbb{E}(T_i^h) + 1\\

\boxed{\frac{1}{2}\mathbb{E}\left((\tilde{Z}^{(n)}_t-\tilde{Z}^{(n)}_{t+h})^2\right) = 2h}

4. Le vérifier expérimentalement en calculant $\frac1{2\text{card}\{(i,j),|t_i-t_j|=h\}}\sum_{|t_i-t_j|=h}(\tilde{Z}^{n}_{t_i} - \tilde{Z}^{n}_{t_j})^2$, où $\text{card}\{(i,j),|t_i-t_j|=h\}$ représente le nombre de couples $(t_i,t_j)$ distants de $h$, pour $h \in \{0.001,0.002,\ldots,0.3\}$. Représenter graphiquement cette quantité.

```python
n = 200
liste_quantite = []
liste_h_bons = []
for h in np.linspace(0.001,0.30,300) :
    # Détermination de l'ensemble d'indices
    liste_t = [i/n for i in range(n+1)]
    liste_dist = []
    for ti in liste_t :
        for tj in liste_t :
            if abs(ti - tj) == h :
                liste_dist.append((ti,tj))

    # Calcul de la quantité
    card = len(liste_dist)
    somme = 0
    if card != 0 :
        liste_h_bons.append(h)
        for ti, tj in liste_dist :
            Zj = 0 # Calcul de Ztj
            Zi = 0 # Calcul de Zti
            for i in range(n+1) :
                X = np.random.choice([-1,1])
                U = np.random.rand()
                if ti <= U :
                    Zi += X
                else :
                    Zi += -X
                if tj <= U :
                    Zj += X
                else :
                    Zj += -X
            Zi = Zi / np.sqrt(n)
            Zj = Zj / np.sqrt(n)
            somme += (Zi - Zj) ** 2
        quantite = 1 / 2 / card * somme
        liste_quantite.append(quantite)
```

```python
plt.plot(liste_h_bons, liste_quantite, color = "green")
plt.title("Quantité évaluée en fonction de h")
plt.xlabel("h")
plt.ylabel("1/2card * Somme")
plt.grid()
plt.show()
```

On vérifie ainsi expérimentalement que a = 2, il s'agit de la pente de la droite aprochée.

Pour ce faire, nous allons simuler des droites aléatoires, avec la paramétrisation suivante :
$$x \cos(\alpha) + y \sin(\alpha) = p,$$
où $x,y \in [0,1]$ sont les coordonnées, $\alpha \sim \mathcal{U}_{[0,\pi]}$ est la variable d'*orientation*, $p \sim \mathcal{U}_{[-\sqrt{2},\sqrt{2}]}$ est la variable de *position*.

On peut ainsi écrire la version 2D de $\tilde{Z}$ de la manière suivante :
$$\tilde{Z}^{(n)}_{(x,y)} = \frac{1}{\sqrt{n}}\sum_{i=1}^n X_i (1_{\{x \cos(\alpha) + y \sin(\alpha) \leq p\}}-1_{\{x \cos(\alpha) + y \sin(\alpha) > p\}})$$

3. Combiner les résultats des deux questions précédentes pour générer une réalisation de $\tilde{Z}$ sur $[-1,1]^2$ pour $n = 100, 1000$ et $10000$ sur la même grille de discrétisation. Commenter. 

```python
def plot(n, nombres_droites) :
    liste_alpha = []
    liste_p = []
    for i in range(nombres_droites) :
        liste_alpha.append(np.random.rand() * np.pi)
        liste_p.append(np.random.rand() * 2 * np.sqrt(2) - np.sqrt(2))
        
    Z_tield_sum = 0
    xx, yy = np.meshgrid(np.linspace(-1, 1, 200),np.linspace(-1, 1, 200))
    for alpha, p in zip(liste_alpha, liste_p) :
        Z = 0
        for j in range(n+1) :
            X = np.random.choice([-1,1])
            Z += -X
        Z = Z / np.sqrt(n)
        Z_tield = (xx * np.cos(alpha) + yy * np.sin(alpha) <= p) * Z
        Z_tield_sum += Z_tield
    plt.pcolormesh(xx, yy, Z_tield_sum, shading="gouraud")
    plt.colorbar()
    plt.title(f"Marche aléatoire en 2 dimensions\npour n = {n} et {nombres_droites} droites")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```
